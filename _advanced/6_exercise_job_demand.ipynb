```python
import matplotlib.pyplot as plt
import pandas as pd 
from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python

df.info()
```

```python
df_us = df[df['job_country'] == 'United States'].copy()
```

```python
df_us
```

```python
df_us['job_month']= df_us['job_posted_date'].dt.strftime('%B')
```

```python
df_us
```

```python
df_us_pivot = df_us.pivot_table(index = 'job_month', columns='job_title_short', aggfunc='size')
```

```python
df_us_pivot
```

```python
df_us_pivot = df_us_pivot.reset_index()
```

```python
df_us_pivot['month_no']= pd.to_datetime(df_us_pivot['job_month'], format ='%B').dt.month
```

```python
df_us_pivot
```

```python
df_us_pivot = df_us_pivot.sort_values('month_no')
```

```python
df_us_pivot
```

```python
df_us_pivot.drop(columns=['month_no'], inplace=True)
```

```python
df_us_pivot
```

```python
top_countries = df['job_title_short'].value_counts().head(3)
top_countries = top_countries.index.tolist()
top_countries
```

```python
df_us_pivot.set_index('job_month')
```

```python
df_us_pivot
```

```python
df_us_pivot.set_index('job_month', inplace=True) 
```

```python
df_us_pivot
```

```python
df_us_pivot[top_countries].plot(kind='line')
plt.ylabel('Number of Job Postings')        
plt.title('Job Demand in the US by Month for Different Job Titles')     
plt.legend(title='Job Title', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
