```python
import matplotlib.pyplot as plt
import pandas as pd 
from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df_r = pd.read_csv('https://lukeb.co/software_csv')
```

```python
df_us = df[df['job_country'] == 'United States'].copy()
df_us
```

```python
df_us['job_month'] = df_us['job_posted_date'].dt.strftime('%B')
```

```python
df_us_pivot = df_us.pivot_table(index='job_month', columns='job_title_short', aggfunc='size')
```

```python
df_us_pivot = df_us_pivot.reset_index()
df_us_pivot
```

```python
df_us_pivot['month_no'] = pd.to_datetime(df_us_pivot['job_month'], format='%B').dt.month
df_us_pivot 
```
