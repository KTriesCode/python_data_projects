```python
import matplotlib.pyplot as plt
import pandas as pd 
from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

Pivot table method df.pivot_table


```python
 df.pivot_table(index ='job_title_short', aggfunc='size').sort_values(ascending=False).head(10).plot(kind='barh', figsize=(10,6), title='Top 10 Job Titles')
```

```python
df_country_salary = df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
df_country_salary.head(1)
```

```python
top_countries = df['job_country'].value_counts().head(6).index
top_countries
df_top_countries = df_country_salary.loc[top_countries]
df_top_countries
```

```python
job_title = ['Data Scientist', 'Data Engineer', 'Data Analyst']
df_top_countries = df_top_countries[job_title] 
df_top_countries
```

```python
df_top_countries.plot(kind='bar', figsize=(10,6), title='Median Salary by Job Title in Top 6 Countries')
plt.ylabel('Median Salary (USD)')
plt.xlabel('Country')       
plt.xticks(rotation=45)
plt.legend(title='Job Title')
plt.show()
```
